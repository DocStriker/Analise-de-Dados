```python
#importar as principais bibliotecas para análise

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

#tirando os avisos do pandas
warnings.filterwarnings('ignore')
```

```python
#tendo uma visualização geral das primeiras 5 linhas da base de dados
bd = pd.read_csv('Salary_Data.csv')
bd.head()
```

```python
#renomeando as colunas
bd.rename(columns = {'YearsExperience' : 'Xp', 'Salary' : 'Renda'}, inplace= True)
bd.head()
```

```python
bd.shape
```

```python
bd.isnull().sum()
```

```python
#verificando se há campos nulos
sns.heatmap(bd.isnull(), cbar=False);
```

```python
#analisando as principais informações estatísticas da base de dados
bd.describe()
```

```python
#plotando um gráfico de densidade com o seaborn kdeplot
plt.figure( figsize=(10 , 6))
plt.title('Análise de Renda')

sns.kdeplot(bd['Renda'], color='g');
```

```python
plt.figure( figsize=(10 , 6))
plt.title('Análise da Xp')

sns.kdeplot(bd['Xp'], color='b');
```

```python
#plotando um boxplot para análise das médias
sns.boxplot(bd['Renda']);
```

```python
sns.boxplot(bd['Xp']);
```

```python
#fazendo a plotando de um gráfico de disperção
sns.scatterplot(data=bd, x='Renda', y='Xp');
```

```python
#verificando a regressão linear do gráfico e das informações da base de dados
sns.regplot(data=bd, x='Renda', y='Xp');
```

```python
#verificando a correlação entre a coluna de Renda e Xp
Correlacao = bd.corr('pearson')
sns.heatmap(Correlacao, annot=True);
```

```python
Correlacao = bd.corr('spearman')
sns.heatmap(Correlacao, annot=True);
```
