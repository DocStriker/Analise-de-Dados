```python
# Bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
# Lendo os dados
bd = pd.read_csv('Dados_Indendio.csv', encoding='latin-1')

# Verificando
bd.head()
```

```python
# Nulos
bd.isnull().sum()
```

```python
# Nulos Plot
plt.figure(figsize=(14,5))
plt.title('An√°lise de campos nulos')
sns.heatmap(bd.isnull(), cbar=False);
```

```python
# Estat√≠sticas
bd.describe()
```

```python
# Info
bd.info()
```

```python
# Campos √∫nicos
bd.nunique()
```

```python
# analise por ano dos inc√™ndios
Analise = bd.groupby(by=['year']).sum(numeric_only=True).reset_index()

Analise.head()

# Tamanho
plt.figure(figsize=(12,5))

# Style
plt.style.use('ggplot')

# Gr√°fico
plt.title('Total inc√™ndios no Brasil 1997 - 2017', loc='left', fontsize=14)
sns.lineplot( data=Analise, x='year', y='number', estimator='sum', lw=2, color='#ff5555', alpha=0.85)

# Labels
plt.xlabel('Quantidade')
plt.ylabel('Per√≠odo');
```

```python
Analise_02 = bd.groupby(by=['year', 'month']).sum(numeric_only=True).reset_index()
Analise_02.head()

# Tamanho
plt.figure(figsize=(12,5))

# Gr√°fico
plt.title('Inc√™ndios por m√™s', loc='left', fontsize=14)
sns.boxplot(data=Analise_02, x='month', y='number', palette='coolwarm', saturation=1, width=0.9, linewidth=2,
           order=['Janeiro', 'Fevereiro', 'Mar√ßo', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'])

# Labels
plt.xlabel('M√™s')
plt.ylabel('N√∫mero de inc√™ndios');
```

```python
Analise_03 = bd.groupby(by=['state']).sum(numeric_only=True).reset_index()[['state', 'number']].sort_values('number', ascending=False)
Analise_03.head()

# Tamanho
plt.figure(figsize=(12,5))

# Gr√°fico
plt.title('Estados com maior n√∫mero de inc√™ndios', loc='left', fontsize=14)
plt.bar(Analise_03.state, Analise_03['number'], color='#f44e3f')

# Labels
plt.xticks(rotation=90)
plt.ylabel('Quantidade');
```

```python
Lista_TOP10 = Analise_03['state'][0:10].values

# Tamanho
plt.figure(figsize=(12,5))

# Gr√°fico
plt.title('Top 10 Estados com inc√™ndios', loc='left', fontsize=14)

# Loop
for Coluna in Lista_TOP10:
    # Filtrar o estado
    Filtro = bd.loc[bd['state'] == Coluna]
    
    # Agrupar os valores para sumarizar
    Analise_Local = Filtro.groupby(by=['year']).sum(numeric_only=True).reset_index()
    
    # Plot
    sns.lineplot(data=Analise_Local, x='year', y='number', lw=2, alpha=0.85)

# Labels
plt.xlabel('Per√≠odo')
plt.ylabel('N√∫mero de inc√™ndios')

# Legenda
plt.legend(Lista_TOP10, bbox_to_anchor=(1, 0.7));
```
