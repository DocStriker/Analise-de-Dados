```python
# Bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
# lendo a base de dados
bd = pd.read_excel('Dados_Pib.xlsx')
# Verificando
bd.head(10)
```

```python
# Agrupando a base de dados por território
bd.groupby(by=['Territorialidades', 'Ano']).mean()
```

```python
# Sistema de Grids

# Cor de fundo
backg = '#f5f5f5'

# Criar o sistema de Grids
Grid_Graf = sns.FacetGrid(bd, col='Territorialidades', hue='Territorialidades', col_wrap=4)

# Adicionar gráfico de linhas em cada gráfico
Grid_Graf = Grid_Graf.map(plt.plot, 'Ano', 'PIB per capita')

# Adicionar uma sombra + ajuste do título
Grid_Graf = Grid_Graf.map(plt.fill_between, 'Ano', 'PIB per capita', alpha=0.2).set_titles('{col_name} Territorialidades')

# Filtro do título
Grid_Graf = Grid_Graf.set_titles('{col_name}')

# Adicionar um subtítulo
Grid_Graf = Grid_Graf.fig.suptitle(                                
    'Evolução da Renda per capita por Estado \n Esse relatório foi elaborado no treinamento "Python para Análise de Dados" \n Está disponível no canal do youtube @Data Viking',
    fontsize=18
)

# Ajustando
plt.subplots_adjust(top=0.92)
```
