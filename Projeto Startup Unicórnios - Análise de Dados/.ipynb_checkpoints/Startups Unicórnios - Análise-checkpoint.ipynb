```python
#importando as bibliotecas para o uso na análise do banco de dados
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
#importando e visualizando as primeiras 5 linhas do banco de dados 'Startups-Unicórnios'
BD = pd.read_csv('unicorns till sep 2022.csv')
BD.head()
```

```python
#analisando a quantidade de registros
BD.shape
```

```python
#analisando os nomes de suas colunas
BD.columns
```

```python
#renomeando as colunas para pt-br
BD.rename(columns={
    'Company': 'Empresa',
    'Valuation ($B)': 'Valor ($)',
    'Date Joined': 'Data de Adesão',
    'Country': 'País',
    'City': 'Cidade',
    'Industry': 'Setor',
    'Investors': 'Investidores'
}, inplace=True)
BD.columns
```

```python
#analisando as informações gerais do banco de dados
BD.info()
```

```python
#verificando a quantidades de campos nulos no banco de dados por coluna
BD.isnull().sum()
```

```python
#plotando um mapa de calor para analisar os campos nulos
plt.figure(figsize=(15,6))
plt.title('Analisando Campos Nulos')
sns.heatmap(BD.isnull(), cbar=False);
```

```python
#verificando a quantidade de informações únicas em cada coluna no banco de dados
BD.nunique()
```

```python
#verificando o todos as informações únicas da coluna 'Setor'
BD['Setor'].unique()
```

```python
#analisando a porcetagem de ocupação de cada setor no banco de dados
round(BD['Setor'].value_counts(normalize=True) * 100, 1)
```

```python
#plotando a análise da porcentegem dos setores no banco de dados em um gráfico de barras
plt.figure(figsize=(15,6))
plt.title('Análise dos Setores')
plt.bar(BD['Setor'].value_counts().index, BD['Setor'].value_counts())
plt.xticks(rotation=45, ha='right');
```

```python
#analisando a porcentagem dos países que possuem startups unicórnios no banco de dados
Analise = round(BD['País'].value_counts(normalize=True) * 100, 1)
Analise
```

```python
#plotando essa análise em um gráfico de pizza
plt.figure(figsize=(15,6))
plt.title('Análise dos Países - Top 5', pad=60)
plt.pie(
    Analise.head(),
    labels= Analise.index[0:5],
    rotatelabels=45,
    shadow=True,
    startangle=90,
    autopct='%1.1f%%',
    pctdistance=0.8
);
```

```python
#convertendo os valores na coluna 'Valor ($)' para tipo numérico
BD['Valor ($)'] = pd.to_numeric(BD['Valor ($)'].apply(lambda Linha: Linha.replace('$', '')))
```

```python
#visualizando as 5 primeiras linhas dessa coluna para verificar a conversão
BD['Valor ($)'].head()
```

```python
#convertendo a coluna 'Data de Adesão' para o tipo 'datetime64' e visualizando essa conversão
BD['Data de Adesão'] = pd.to_datetime(BD['Data de Adesão'])
BD['Data de Adesão'].head()
```

```python
#criando duas novas colunas separando o mês e o ano da coluna 'Data de Adesão' e gerando uma tabela
#por 'País', 'Ano', 'Mês' e 'Empresa' e analisando a data das primeiras startups que viraram unicórnios
BD['Mês'] = pd.DatetimeIndex(BD['Data de Adesão']).month
BD['Ano'] = pd.DatetimeIndex(BD['Data de Adesão']).year
Analise_Agrupada = BD.groupby(by=['País','Ano','Mês','Empresa']).count()['Setor'].reset_index()
Analise_Agrupada.sort_values('Ano')
```

```python
#analisando as primeiras startups unicórnios no Brasil
Analise_Agrupada.loc[Analise_Agrupada['País'] == 'Brazil']
```

```python
#analizando o resumo do banco de dados com as novas colunas e conversões de tipos de dados
BD.info()
```

```python
#fazendo uma tabela de ranking com o valor total de startups em dolar por país
Analise_Valor = BD.groupby(by=['País']).sum(numeric_only=True)['Valor ($)']\
.reset_index().sort_values('Valor ($)', ascending=False).reset_index(drop=True)
Analise_Valor
```

```python
#plotando um gráfico de curvas para essa análise
plt.figure(figsize=(15,6))
plt.plot(Analise_Valor['País'], Analise_Valor['Valor ($)'])
plt.xticks(rotation=45, ha='right');
```
