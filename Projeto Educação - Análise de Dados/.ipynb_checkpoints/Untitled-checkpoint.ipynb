```python
g = sns.JointGrid(marginal_ticks=True, space=0.5).refline(x=60, y=60)
x, y = bd['math score'], bd['writing score']
sns.scatterplot(x=x, y=y, ax=g.ax_joint, hue=bd['gender'])
sns.histplot(x=x, ax=g.ax_marg_x, hue=bd['gender'])
sns.kdeplot(y=y, ax=g.ax_marg_y, hue=bd['gender'])
```
