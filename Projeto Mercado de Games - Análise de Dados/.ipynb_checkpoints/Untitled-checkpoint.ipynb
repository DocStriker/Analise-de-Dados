```python
#importando as principais bibliotecas para an√°lise de dados

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
#abrindo o arquivo com os dados e visualizando as 5 primeiras linhas
bd = pd.read_csv('PS4_GamesSales.csv', encoding='latin-1')
bd.head()
```

```python
#visualizando a quantidade de dados em nosso dataset
bd.shape
```

```python
#verificando a quantidade de campos nulos na base de dados
bd.isnull().sum()
```

```python
#visualizando em forma de gr√°fico
plt.figure(figsize=(16,5))
sns.heatmap(bd.isnull(), cbar=False);
```

```python
#exclu√≠ndo os campos nulos
bd.dropna(inplace=True)
```

```python
bd.isnull().sum()
```

```python
#visualizando um resumo estat√≠stico da base dados
bd.describe()
```

```python
#visualizando a vendas total de unidades globais por ano
sns.barplot(data=bd, x='Year',y='Global', ci=None, color='#69b3a2', estimator=sum);
```

```python
bd = bd.loc[(bd['Year'] != 2019) & (bd['Year'] != 2020)]
bd.head()
```

```python
#visualizando a concentra√ß√£o dos dados em rela√ß√£o a quantidade de vendas de unidades
plt.figure(figsize=(12,5))
plt.style.use('ggplot')
sns.kdeplot(bd['Global'], shade=True, bw=1, color='#96a8a8', linewidth=2.5);
```

```python
Analise = bd.groupby(by=['Year']).sum(numeric_only=True)
Analise
```

```python
#plotando um gr√°fico boxplot para visualizar o quantidade de unidades vendidas por ano
plt.figure(figsize=(10,4))
sns.boxplot(data=bd, x='Year', y='Global');
```

```python
bd.loc[bd['Global'] >= 10]
```

```python
#fazendo uma rela√ß√£o de 100%
America = [America / Total * 100 for America, Total in zip(Analise['North America'], Analise['Global'])]
Europa = [Europa / Total * 100 for Europa, Total in zip(Analise['Europe'], Analise['Global'])]
Japao = [Japao / Total * 100 for Japao, Total in zip(Analise['Japan'], Analise['Global'])]
Mundo = [Mundo / Total * 100 for Mundo, Total in zip(Analise['Rest of World'], Analise['Global'])]

America, Europa, Japao, Mundo
```

```python
#plotando a rela√ß√£o de 100%
plt.figure(figsize=(10,5))

Largura_Barra = 0.85
Rotulos = Analise.index
Grupos = [0, 1, 2, 3, 4, 5]

plt.bar(Grupos, America, width=Largura_Barra, color='#b5ffb9', edgecolor='white')
plt.bar(Grupos, Europa, bottom=America, width=Largura_Barra, color='#f9bc86', edgecolor='white')
plt.bar(Grupos, Japao, bottom=[A + B for A, B in zip(America, Europa)], width=Largura_Barra, color='#a3acff', edgecolor='white')
plt.bar(Grupos, Mundo, bottom=[A + B + C for A, B, C in zip(America, Europa, Japao)], width=Largura_Barra, color='#d3acfe', edgecolor='white')

plt.xticks(Grupos, Rotulos)
plt.xlabel('Grupo')
plt.ylabel('Distribui√ß√£o %')

plt.legend(['America N', 'Europa', 'Jap√£o', 'Resto do Mundo'], loc='upper left', bbox_to_anchor=(0.15, -0.1), ncol=4);
```

```python
#verificando os valores √∫nicos da coluna 'Publisher'
bd['Publisher'].nunique()
```

```python
#importando a biblioteca sklearn de machine learning para classificar as informa√ß√µes de texto em n√∫meros
from sklearn.preprocessing import LabelEncoder

Funcao_Label = LabelEncoder()
bd['Produtor'] = Funcao_Label.fit_transform(bd['Publisher'])
bd['Genero'] = Funcao_Label.fit_transform(bd['Genre'])
bd['Jogo'] = Funcao_Label.fit_transform(bd['Game'])
bd.head()
```

```python
#criando uma paleta de cores com 8 cores diferentes
Paleta_Cores = sns.color_palette('husl', 8)
Paleta_Cores
```

```python
#relacionando a quantidade de unidades vendidas com as empresas que fizeram os games
plt.figure(figsize=(20,5))
plt.axhline(y=10, linestyle='-', color='g', label='Linha Horizontal')
sns.scatterplot(data=bd, x='Produtor',  y='Global', color=Paleta_Cores[0])
```

```python
bd.loc[bd['Global'] >= 10, ['Publisher', 'Produtor', 'Global']]
```

```python
#relacionando a quantidade de unidades vendidas com o tipo de g√™nero dos games
plt.figure(figsize=(20,5))
plt.axhline(y=10, linestyle='-', color='g', label='Linha Horizontal')
sns.scatterplot(data=bd, x='Genero',  y='Global', color=Paleta_Cores[0])
```

```python
bd.loc[bd['Global'] >= 10, ['Genre', 'Genero', 'Global']]
```

```python
#relacionando a quantidade de unidades vendidas com os nomes dos games
plt.figure(figsize=(20,5))
plt.axhline(y=10, linestyle='-', color='g', label='Linha Horizontal')
sns.scatterplot(data=bd, x='Jogo',  y='Global', color=Paleta_Cores[0])
```

```python
bd.loc[bd['Global'] >= 10, ['Game', 'Jogo', 'Global']]
```
