```python
# Bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
# Lendo a base de dados
bd = pd.read_excel('Dados_Empresas_Energia.xlsx')

# Verificando
bd.head()
```

```python
# Series Temporais - Setar o Index
bd.set_index('Data', inplace=True)
bd.head()
```

```python
# Grfico

# Estilo do grfico
plt.style.use('seaborn-darkgrid')

# Tamanho
plt.figure(figsize=(16,7))

# Ttulo
plt.title('Anlise aes de empresas de energia', loc='left', fontsize=18, fontweight=0)

# Plot da Petrobras
plt.plot(bd.index, bd['Petrobras'], color='#008c4a', linewidth=4, alpha=0.7)
# Texto da Petrobras
plt.text(bd.index[-1], bd['Petrobras'].tail(1), 'Petrobras', color='#008c4a', size='large', horizontalalignment='left')

# Plot de todas as colunas
for Coluna in bd.columns[1:]:
    # Plot das outras aes
    plt.plot(bd.index, bd[Coluna], color='gray', linewidth=2, alpha=0.7)
    # Texto das outras aes
    plt.text(bd.index[-1], bd[Coluna].tail(1), Coluna, color='gray', size='large', horizontalalignment='left')

# Labels
plt.xlabel('Perodo')
plt.ylabel('Preo de Fechamento ($)');
```
